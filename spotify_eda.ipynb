```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('/content/tracks.csv')
df_tracks.head()
```

```python
#null values
print(df_tracks.isnull().sum())
```

```python
df_tracks.info()
```

```python
#least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending=True)
sorted_df.head(10)
```

```python
df_tracks.describe().transpose()
```

```python
#most popular
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
# Reset the index to restore 'release_date' as a column
df_tracks.reset_index(inplace=True)

# Now you can set the index
df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index, format='%Y', errors='coerce') # Handle different date formats and invalid dates
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
# Identify non-numeric columns
non_numeric_cols = df_tracks.select_dtypes(exclude=['number']).columns

print(non_numeric_cols)

#Drop non-numeric columns before calculating correlation
corr_df=df_tracks.drop(non_numeric_cols,axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='r').set(title='Popularity vs Acousticness Correlation')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='No. of Tracks per Year')
```

```python
total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
# Group by year and calculate the mean duration for each year to avoid duplicate labels
year_groups = df_tracks.groupby(years)['duration'].mean()
fig = sns.barplot(x=year_groups.index, y=year_groups.values, ax=ax, err_kws={'linewidth': False}).set(title='year vs duration')
plt.xticks(rotation=90)
```

```python
# Reset the index of df_tracks to ensure unique index values
df_tracks = df_tracks.reset_index()

# Extract the year and duration
years = df_tracks['release_date'].dt.year
total_dr = df_tracks['duration']

sns.set_style(style='darkgrid')
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
# Group by year and calculate the mean duration for each year to avoid duplicate labels
year_groups = df_tracks.groupby(years)['duration'].mean()
fig = sns.lineplot(x=year_groups.index, y=year_groups.values, ax=ax).set(title='year vs duration')
plt.xticks(rotation=60)
plt.show()
```

```python
#analysis on genre
df_genre=pd.read_csv('/content/SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.title('Distribution of songs in different genres')
sns.color_palette("rocket",as_cmap= True)
sns.barplot(y='genre', x='duration_ms', data= df_genre)
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
```

```python
#top 5 based on popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data= famous).set(title='Top 5 genres by Popularity')
plt.xticks(rotation=90)
```
